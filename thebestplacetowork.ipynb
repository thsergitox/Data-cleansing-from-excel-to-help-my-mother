```python
import pandas as pd
```

```python
#Here we read the .xlsd where there are all the job options for my mom
df = pd.read_excel("PLAZA VACANTES 2023 PRIMARIA.xlsx, skiprows = 2, nrows = 239)
```

```python
df.head()
```

```python
#To recognize all the columns
print(df.columns.values)
```

```python
#Data clensing function
def manipulation(df, name): 
    #We see if it is rural or not
    if name == 'rurales':
        df = df[df["RURAL"].notnull()]
    else: 
        df = df[df["RURAL"].isna()]
    #My mom is a proffesor
    df = df[df["CARGO"] == "PROFESOR"]
    #My mom does not work as a sports teacher
    df = df[df["AREA CURRICULAR/ CAMPO DE CONOCIMIENTO"] != "EDUCACION FISICA"]
    #My mom does not speak native languages
    df = df[df["LENGUA ORIGINARIA"].isna()]
   	#The best option for job
    df = df[df["TIPO DE PLAZA"] == "ORGANICA"]
    #Creating a new excel for the rural options and non rural options
    df.to_excel(name + ".xlsx", index = False)
    return df
```

```python
#Here I chose the rural options
rural = df
rural = manipulation(rural, "rurales")
rural.head()
```

```python
#And here the non rural options
norural = df
norural = manipulation(norural, "norurales")
norural.head()
```
